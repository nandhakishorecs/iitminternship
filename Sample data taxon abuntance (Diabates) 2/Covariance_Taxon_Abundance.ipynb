```python
#extracting Genus Name
df['GenusName']= df['Unnamed: 0'].apply(lambda x: str(x).replace(';', ', '))

df['GenusName']=df['GenusName'].str.split().str[-3:-2]
df['GenusName']=df['GenusName'].apply(lambda x: str(x).replace(',',''))

genus=df['GenusName']
df=df.drop(columns=['GenusName'])

df.insert(1,'Genus',genus)
df['Genus']=df['Genus'].apply(lambda x: str(x).replace('\'',''))
```

```python
#extracting Family Name
df['FamilyName']= df['Unnamed: 0'].apply(lambda x: str(x).replace(';', ', '))

df['FamilyName']=df['FamilyName'].str.split().str[-4:-3]
df['FamilyName']=df['FamilyName'].apply(lambda x: str(x).replace(',',''))

family=df['FamilyName']
df=df.drop(columns=['FamilyName'])

df.insert(1,'Family',family)
df['Family']=df['Family'].apply(lambda x: str(x).replace('\'',''))
```

```python
species_df=species_df.drop(['Species'],axis=1)
covarianceMatrix=species_df.corr(method='pearson')

covarianceMatrix
```

```python
for row in covarianceMatrix: 
    for col in covarianceMatrix:
        if(covarianceMatrix[row][col]>=0.5):
            covarianceMatrix[row][col]=1
        else:
            covarianceMatrix[row][col]=0

covarianceMatrix
```

```python
import networkx as nx
G=nx.Graph(covarianceMatrix, with_labels=False)

three_cliques=[]
four_cliques=[]
five_cliques=[]
six_cliques=[]
seven_cliques=[]
eight_cliques=[]
nine_cliques=[]
ten_cliques=[]
eleven_cliques=[]
twelve_cliques=[]
remaining_cliques=[]
cliques=nx.enumerate_all_cliques(G)
for item in cliques: 
    #print(item)
    if(len(item)==3):
        three_cliques.append(item)
    elif(len(item)==4):
        four_cliques.append(item)
    elif(len(item)==5):
        five_cliques.append(item)
    elif(len(item)==6):
        six_cliques.append(item)
    elif(len(item)==7):
        seven_cliques.append(item)
    elif(len(item)==8):
        eight_cliques.append(item)        
    elif(len(item)==9):
        nine_cliques.append(item)
    elif(len(item)==10):
        ten_cliques.append(item)
    elif(len(item)==11):
        eleven_cliques.append(item)
    elif(len(item)==12):
        twelve_cliques.append(item)
    else:
        remaining_cliques.append(item)



print('Number of three cliques:\t',len(three_cliques))
print('Number of four cliques:\t',len(four_cliques))
print('Number of five cliques:\t',len(five_cliques))
print('Number of six cliques:\t',len(six_cliques))
print('Number of seven cliques:\t',len(seven_cliques))
print('Number of eight cliques:\t',len(eight_cliques))
print('Number of nine cliques:\t',len(nine_cliques))
print('Number of ten cliques:\t',len(ten_cliques))
print('Number of eleven cliques:\t',len(eleven_cliques))
print('Number of twelve cliques:\t',len(twelve_cliques))
print('Number of other cliques:\t',len(remaining_cliques))
```

```python
from matplotlib import pylab
import networkx as nx

def save_graph(graph,file_name):
    #initialze Figure
    plt.figure(num=None, figsize=(500, 500), dpi=200)
    plt.axis('off')
    fig = plt.figure(1)
    pos = nx.spring_layout(graph)
    nx.draw_networkx_nodes(graph,pos)
    nx.draw_networkx_edges(graph,pos)
    #nx.draw_networkx_labels(graph,pos)

    cut = 1.00
    xmax = cut * max(xx for xx, yy in pos.values())
    ymax = cut * max(yy for xx, yy in pos.values())
    plt.xlim(0, xmax)
    plt.ylim(0, ymax)

    plt.savefig(file_name,bbox_inches="tight")
    pylab.close()
    del fig

#Assuming that the graph g has nodes and edges entered
save_graph(G,"my_graph.png")

#it can also be saved in .svg, .png. or N .ps formats
```
