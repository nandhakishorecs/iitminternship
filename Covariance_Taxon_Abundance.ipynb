```python
# Correcting Typo errors in the name of the micro organism

df.at[14, 'Unnamed: 0']='Bacteria;Actinobacteria;Actinobacteria;Actinomycetales;Actinomycetaceae;Actinomyces;Actinomyces sp.oral_taxon_181'
df.at[188, 'Unnamed: 0']='Bacteria;Firmicutes;Bacilli;Lactobacillales;Enterococcaceae;Enterococcus;Enterococcus sp.RIT_PI_f'
df.at[472, 'Unnamed: 0']='Bacteria;Firmicutes;Negativicutes;Selenomonadales;Selenomonadaceae;Selenomonas;Selenomonas sp.oral_taxon_892'
df.at[586, 'Unnamed: 0']='Bacteria;Proteobacteria;Gammaproteobacteria;Pasteurellales;Pasteurellaceae;Aggregatibacter;Aggregatibacter sp.oral_taxon_458'
```

```python
# Splitting the columns and stored it in separate dataframe and can take the desired column
df_col = df['Unnamed: 0'].str.split(';', n=6, expand =True)
```

```python
taxa_df["Kingdom"] = pd.DataFrame(df_col[0])
taxa_df["Phylum"] = df_col[1]
taxa_df["Phylum"] = df_col[1]
taxa_df["Class"] = df_col[2]
taxa_df["Order"] = df_col[3]
taxa_df["Family"] = df_col[4]
taxa_df["Genus"] = df_col[5]
taxa_df["Species"]=df_col[6]
taxa_df.head()
```

```python
#extracting Species Name
df['SpeciesName']= df['Unnamed: 0'].apply(lambda x: str(x).replace(';', ', '))

df['SpeciesName']=df['SpeciesName'].str.split().str[-2:]

species=df['SpeciesName']
df=df.drop(columns=['SpeciesName'])

df.insert(1,'Species',species)
```

```python
#extracting Genus Name
df['GenusName']= df['Unnamed: 0'].apply(lambda x: str(x).replace(';', ', '))

df['GenusName']=df['GenusName'].str.split().str[-3:-2]
df['GenusName']=df['GenusName'].apply(lambda x: str(x).replace(',',''))

genus=df['GenusName']
df=df.drop(columns=['GenusName'])

df.insert(1,'Genus',genus)
```

```python
# Cleaning the unnecessary punctutaions in the name of the micro organism 
df['Genus']=df['Genus'].apply(lambda x: str(x).replace('\'',''))
df['Species']=df['Species'].apply(lambda x: str(x).replace(',',''))
df['Species']=df['Species'].apply(lambda x: str(x).replace('\'',''))
```

```python
new_df = df.copy()

# Adding Genus & Species to the dataframe
new_df.insert(0,'Species',taxa_df['Species'])
new_df.insert(0,'Genus',taxa_df['Genus'])
new_df.head()
```

```python
new_df.to_csv('newDF.csv')
```

```python
#replaceing NaN with mean of their columns 
values_df=new_df.drop(columns=['Unnamed: 0','Genus','Species'], axis=1)
```

```python
values_df.isnull().values.any()
values_df.isnull().sum()
```

Here, NaN represents either absence of the micro-organisms due to technical or biological problem. So we can't impute the NaNs using mean. We can replace the NaN with **zero**.

```python
values_df=values_df.fillna(0)
```
