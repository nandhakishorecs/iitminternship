```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Original_data_frame=pd.read_csv('DIABIMMUNE_WGS.WGS.taxon_abundance.tsv', sep='\t')

df=Original_data_frame.copy()
```

```python
#extracting Species Name
df['SpeciesName']= df['Unnamed: 0'].apply(lambda x: str(x).replace(';', ' '))


df['SpeciesName']=df['SpeciesName'].str.split().str[-2:]

species=df['SpeciesName']
df=df.drop(columns=['SpeciesName'])

df.insert(1,'Species',species)
```

```python
df=df.drop(columns=['Unnamed: 0'])
df
```
